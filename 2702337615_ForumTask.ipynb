```python
# Forum Task 1 :D

lloyd = {'name' : "Lloyd", "homework" : [], "quizzes" : [], "tests" : []}
alice = {'name' : "Alice", "homework" : [], "quizzes" : [], "tests" : []}
tyler = {'name' : "Tyler", "homework" : [], "quizzes" : [], "tests" : []}

lloyd = {
"name": "Lloyd",
"homework": [90.0,97.0,75.0,92.0],
"quizzes": [88.0,40.0,94.0],
"tests": [75.0,90.0]
}
alice = {
"name": "Alice",
"homework": [100.0, 92.0, 98.0, 100.0],
"quizzes": [82.0, 83.0, 91.0],
"tests": [89.0, 97.0]
}
tyler = {
"name": "Tyler",
"homework": [0.0, 87.0, 75.0, 22.0],
"quizzes": [0.0, 75.0, 78.0],
"tests": [100.0, 100.0]
}

students = [lloyd, alice, tyler]

for student in students:
    print(f"Name: \t \t {student['name']}")
    print(f"Homework: \t {student['homework']}")
    print(f"Quizzes: \t {student['quizzes']}")
    print(f"Tests: \t \t {student['tests']}\n")

def average(numbers):
    total = float(sum(numbers))
    average = total / len(numbers)
    return average

def get_average(student):
    homework = average(student['homework'])
    quizzes = average(student['quizzes'])
    tests = average(student['tests'])
    weight_hw = homework * 0.1
    weight_q = quizzes * 0.3
    weight_t = tests * 0.6
    weight_avg = weight_hw + weight_q + weight_t
    return weight_avg

print(f"Lloyd's Average: \t{get_average(lloyd)}")

def get_letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

print(f"Lloyd's Letter Grade: \t{get_letter_grade(get_average(lloyd))}\n")

def get_class_average(students):
    results = []
    for s in students:
        results.append(get_average(s))
    return average(results)

print(f"Class Average: \t\t{get_class_average(students)}")
print(f"Class Letter Grade: \t{get_letter_grade(get_class_average(students))}")
```

```python
# Forum Task 3 :D

days = ( "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday")
sales = []

for day in days:
    while True:
        try: 
            sales_data = float(input(f"Enter sales for {day}: "))
            if sales_data < 0:
                print(f"Input was invalid. Re-enter the sales for {day}: ")
            else:
                sales.append(sales_data)
                break
        except ValueError:
            print(f"Input was invalid. Please re-enter the sales for {day}: ")

total = sum(sales)
minimum_sales = min(sales)
maximum_sales = max(sales)

f_total = "{:,.2f}".format(total)
f_minimum_sales = "{:,.2f}".format(minimum_sales)
f_maximum_sales = "{:,.2f}".format(maximum_sales)

print(f"The total sales for the week is ${f_total}")
print(f"The minimum sales amount was: ${f_minimum_sales}")
print(f"The maximum sales amount was: ${f_maximum_sales}")
```
